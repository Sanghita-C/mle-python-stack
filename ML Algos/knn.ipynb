```python
from collections import Counter
```

```python
class KNN:
    def __init__(self):
        self.x = None
        self.y = None
    def train(self,x,y):
        self.x = x
        self.y = y 
        
    def predict_regression(self,x,k):
       distance = [ (self.distance(x,x_train), y_train) for x_train, y_train in zip(self.x,self,self.y)]
       neighbors = sorted(distance)[:k]
       return sum(label for _,label in neighbors)/k
    
    def predict_classification(self,x,k):
        distance = [(self.distance(x,x_train),y_train) for x_train, y_train in zip(self.x,self.y)]
        neighbors = sorted(distance)[:k]
        return Counter(neighbors).most_common(1)[0][0]

    
```

# How to choose K

K is a hyperparameter whose value is sent by user. K is any positive number. when K = 1, the prediction will be very arbitrary (just dependent of the most nearest neighbor)

when k is too large example close to the number of datapoints, that is not a good estimate too - this will lead us to almost ignore similarity patterns

so how to choose K ?
- one thing could be k = sqrt(number of datapoints)
- another approach (better) is cross validation 

# Cross Validation

- We divide the training dataset into N groups
- We list a range of K to choose from K = [1,2,3....Kmax]
- Now we do the training using the N-1 groups of training dataset and do the validation using the last training group
- We can repeat doing the same thing using each of the N groups for validation
