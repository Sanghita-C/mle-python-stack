```python







import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Let's create a simple example dataset
np.random.seed(42)
house_size = np.random.normal(loc=1500, scale=300, size=100)
num_rooms = house_size / 300 + np.random.normal(0, 0.5, 100)

# Combine into a data matrix
X = np.vstack((house_size, num_rooms)).T

# Plot the original data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.5, label='Original Data')
plt.xlabel('House Size')
plt.ylabel('Number of Rooms')
plt.legend()
plt.title('Data Before PCA')
plt.show()

# Now apply PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Transform back to 2D to visualize the principal component direction
X_recovered = pca.inverse_transform(X_pca)

# Plot the PCA-transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.2, label='Original Data')
plt.scatter(X_recovered[:, 0], X_recovered[:, 1], alpha=0.8, color='red', label='Data after PCA')
plt.xlabel('House Size')
plt.ylabel('Number of Rooms')
plt.legend()
plt.title('Data After PCA')
plt.show()
```
